```python
import matplotlib.pyplot as plt
#matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert
import cv2

import numpy as np
```

```python
ismailnjImage = cv2.imread("image/ismailnj.jpeg")
akuImage = cv2.imread("image/aku.jpeg")

ismailnjCropped = ismailnjImage.copy()
ismailnjCropped = ismailnjCropped[0:256,64:320]

akuCropped = akuImage.copy()
akuCropped = akuCropped[800:1200,400:800]

print('Ismail Ori Shape : ',ismailnjImage.shape)
print('Ismail Crop Shape : ',ismailnjCropped.shape)

print('Camera Ori Shape : ',akuImage.shape)
print('Camera Crop Shape : ',akuCropped.shape)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

# Plot Gambar pertama dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(ismailnjImage, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input 1")

ax[1].imshow(cv2.cvtColor(akuImage, cv2.COLOR_BGR2RGB))
ax[1].set_title('Citra Input 2')
# Plot gambar ketiga dengan cv2.imshow
ismailnjCropped_resized = cv2.resize(cv2.cvtColor(ismailnjCropped, cv2.COLOR_BGR2RGB), (320, 320))
ax[2].imshow(ismailnjCropped_resized)
ax[2].set_title("Citra Output 1")

# Plot gambar keempat dengan cv2.imshow
akuCropped_resized = cv2.resize(cv2.cvtColor(akuCropped, cv2.COLOR_BGR2RGB), (800, 800))

ax[3].imshow(akuCropped_resized)
ax[3].set_title('Citra Output 2')

plt.show()
```

```python
inv = invert(ismailnjCropped)
print('Shape Input : ', ismailnjCropped.shape)
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(cv2.cvtColor(ismailnjImage, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input")

ax[1].hist(ismailnjCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
#membuat salinan dari citra dan mengubah tipenya menjadi float 
copyaku = akuCropped.copy().astype(float)
#menentukan kecerahan
brightness = 100
aku_bright = np.clip(copyaku + brightness, 0, 255).astype(np.uint8)

# Membuat figure subplot  
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
# Plot Gambar Input dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(akuCropped, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input")
# Plot Histogram Input
ax[1].hist(akuCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')
# Plot Gambar Output output
ax[2].imshow(cv2.cvtColor(aku_bright, cv2.COLOR_BGR2RGB))
ax[2].set_title(f"Citra Output (Brightness={brightness})")
# Plot Histogram output
ax[3].hist(aku_bright.ravel(), bins=256, range=(0, 255))
ax[3].set_title('Histogram Output')
```
